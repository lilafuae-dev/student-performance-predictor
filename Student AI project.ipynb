```python
import pandas as pd

data = {
    "hours_studied": [2,3,4,5,6,7,8,2,3,6,7,8,5,4,6,7,3,2,8,9],
    "sleep_hours": [5,6,6,7,7,8,8,5,6,7,8,8,7,6,7,8,6,5,8,9],
    "phone_hours": [6,5,4,3,3,2,2,6,5,3,2,2,3,4,3,2,5,6,2,1],
    "grade": [70,75,78,85,88,92,95,72,76,89,93,96,84,80,90,94,77,71,97,98]
}

df = pd.DataFrame(data)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['hours_studied', 'sleep_hours', 'phone_hours']]
y = df['grade']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
prediction = model.predict([[6, 7, 2]])
print("Predicted grade:", prediction)
```

```python
print("High phone usage:", model.predict([[6,7,6]]))
print("Low phone usage:", model.predict([[6,7,1]]))
```
